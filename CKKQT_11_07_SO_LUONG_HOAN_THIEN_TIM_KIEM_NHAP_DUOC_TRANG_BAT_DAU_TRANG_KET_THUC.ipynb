```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
  
    
# Khởi tạo trình điều khiển Firefox
driver = webdriver.Firefox()
# Mở trang web
url = "https://congkhaiketquathau.moh.gov.vn/Pages/admin/report/TT14/manager/mohVtIvdManager.zul?code=1"
driver.get(url)




# Chờ tải đầy đủ trang web (tuỳ thuộc vào trang web)
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ".row.row-space.z-div")))



# Tìm tất cả các phần tử có class "row row-space z-div"
elements = driver.find_elements(By.CSS_SELECTOR, ".row.row-space.z-div")

# Kiểm tra nếu có phần tử
if elements:
    # Truy cập vào từng phần tử và thực hiện các thao tác
    for index, element in enumerate(elements):
        # Tìm tất cả các phần tử input bên trong phần tử hiện tại
        input_elements = element.find_elements(By.CSS_SELECTOR, "input")

        # Chờ đến khi các phần tử input xuất hiện
        for input_element in input_elements:
            wait.until(EC.visibility_of(input_element))

        # Chờ thêm 1 giây để đảm bảo phần tử tương tác được
        time.sleep(1)

        # Thực hiện các thao tác tương ứng với từng hàng
        if index == 0:
            input_elements[0].clear()
            input_elements[0].send_keys("")  # NHẬP TÊN VẬT TƯ CẦN TÌM KIẾM,ĐANG BỎ TRỐNG
        elif index == 1:
            input_elements[0].clear()
            input_elements[0].send_keys("")  # NHẬP TÊN VẬT TƯ CẦN TÌM KIẾM,ĐANG BỎ TRỐNG
            input_elements[1].clear()
            input_elements[1].send_keys("Bệnh viện Hữu nghị Việt Đức")  # NHẬP TÊN BỆNH VIỆN CẦN LỌC, CHÍNH XÁC

            # CHỌN TÊN BỆNH VIỆN HIỆN RA. Click vào phần tử <i> để mở popup
            i_element = driver.find_element(By.CSS_SELECTOR, 'i.z-bandbox-icon.z-icon-search')
            i_element.click()
            # Chờ TÊN BỆNH VIỆN xuất hiện trong khoảng thời gian tối đa là 3 giây
            popup_div = WebDriverWait(driver, 3).until(
                EC.visibility_of_element_located((By.CSS_SELECTOR, 'div.z-bandbox-popup.z-bandbox-open.z-bandbox-shadow')))
            # Tìm và click vào TÊN BỆNH VIỆN
            div_element = popup_div.find_element(By.CSS_SELECTOR, 'div.z-listcell-content')
            div_element.click()
        elif index == 2:
            input_elements[0].clear()
            input_elements[0].send_keys("")  # NHẬP CHỦNG LOẠI, MÃ SẢN PHẨM,ĐANG BỎ TRỐNG
            input_elements[1].clear()
            input_elements[1].send_keys("")  # NHẬP HÃNG SẢN XUẤT, ĐANG BỎ TRỐTRỐNG
        elif index == 3:
            input_elements[0].clear()
            input_elements[0].send_keys("")  # NHẬP KHOẢNG GIÁ TRÚNG THẦU CẦN TÌM TỪ, ĐANG BỎ TRỐNG
            input_elements[1].clear()
            input_elements[1].send_keys("")  # NHẬP KHOẢNG GIÁ TRÚNG THẦU CẦN TÌM ĐẾN, ĐANG BỎ TRỐNG
            time.sleep(2)
        elif index == 4:
            # TÌM Ô NGÀY CÔNG KHAI
            # Tìm tất cả các phần tử input nested bằng class name
            nested_input_elements = driver.find_elements(By.CLASS_NAME, "z-datebox-input")

            # Lấy phần tử đầu tiên và thực hiện các thao tác
            nested_input_element1 = nested_input_elements[0]
            time.sleep(2)
            nested_input_element1.clear()
            time.sleep(2)
            nested_input_element1.send_keys("01/01/2022")

            # TÌM Ô ĐẾN NGÀY
            # Lấy phần tử thứ hai và thực hiện các thao tác
            nested_input_element2 = nested_input_elements[1]
            nested_input_element2.clear()
            nested_input_element2.send_keys("31/12/2022")

        time.sleep(5)

        
        
        
# Tìm nút "TÌM KIẾM" bằng CSS selector
button = driver.find_element(By.CSS_SELECTOR, 'button.btn-primary.z-button')
# Thực hiện click vào nút "TÌM KIẾM"
button.click()
# Chờ 30 giây để đợi kết quả tải và hiển thị
time.sleep(30)




def wait_for_element_present(driver, by, locator, timeout=10):
    wait = WebDriverWait(driver, timeout)
    wait.until(EC.presence_of_element_located((by, locator)))

    




def get_modal_data():
    # Chờ cho bảng modal hiển thị đủ các phần tử
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "z-window-modal")))

    time.sleep(5)

    # Lấy nội dung HTML của bảng modal
    modal = driver.find_element(By.CLASS_NAME, "z-window-modal")
    html_content = modal.get_attribute("innerHTML")
    soup = BeautifulSoup(html_content, 'html.parser')
    div = soup.find_all('input', class_="z-textbox")[2]
    soluong = div.get('value')

    donvitrungthau = soup.find_all('input', class_="z-textbox")[-2]
    trungthau = donvitrungthau.get('value')

    # Đóng bảng modal
    btn_close = modal.find_element(By.CLASS_NAME, "z-window-close")
    btn_close.click()

    time.sleep(6)

    return soluong, trungthau


def get_data():
    # Lặp qua các liên kết
    a_links = driver.find_elements(By.CLASS_NAME, "z-a")

    i = 0  # Biến đếm số dòng đã lấy được trên mỗi trang

    data_list = []  # Khởi tạo một list để chứa dữ liệu từ trang hiện tại

    for link in a_links:
        i += 1

        print(link)
        print("Text:", link.text)
        try:
            if len(link.text) > 1:
                link.click()
                soluong, donvi = get_modal_data()
                data_list.append([link.text, soluong, donvi])

        except Exception as e:
            print(e)
            continue

    return data_list


def process_page(page_number, rows_to_fetch):
    print("Page", page_number)

    start_row = (page_number - 1) * rows_per_page + 1

    # Lấy dữ liệu từ trang hiện tại
    data_list = get_data()

    # Kiểm tra số dòng lấy được có đúng không
    if len(data_list) < rows_to_fetch:
        print(f"ERROR: Only {len(data_list)} rows fetched instead of {rows_to_fetch} rows.")

    # Tạo tên file Excel cho trang hiện tại
    file_name = f'CKKQT_SL_page_{page_number}.xlsx'

    # Tạo DataFrame từ list dữ liệu
    df = pd.DataFrame(data_list, columns=['Tên Vtyt', 'Số Lượng', 'Đơn vị trúng thầu'])

    # Ghi DataFrame vào file Excel
    try:
        df.to_excel(file_name, index=False)
    except Exception as e:
        print(e)

        

# Tìm ô input page bắt đầu, thay đổi số page bắt đầu ở 2 chỗ

start_page_input = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div/div[2]/div[2]/div/div[2]/div/div[2]/div/div/div[2]/ul/li[3]/input')))
start_page_input.click() # Click vào ô bắt đầu
time.sleep(3)

# Xoá nội dung ô bắt đầu
start_page_input.send_keys(Keys.BACKSPACE)
time.sleep(3)

# Nhập giá trị trang bắt đầu
start_page_input.send_keys("3")
time.sleep(5)
# TỚI ĐÂY DỮ LIỆU SẼ LOAD ĐẾN TRANG SỐ 3 

# Giả lập nhấn phím Enter
actions = ActionChains(driver)
actions.move_to_element(start_page_input).send_keys(Keys.ENTER).perform()
time.sleep(30)


# Bắt đầu lấy dữ liệu từ trang 3 và các trang tiếp theo
page_number = 3  # SỐ TRANG BẮT ĐẦU LẶP
rows_per_page = 50
rows_to_fetch = 50  # Số dòng muốn lấy






while True:
    process_page(page_number, rows_to_fetch)

    # Xử lý lỗi ElementNotInteractableException
    try:
        time.sleep(5)
        next_button = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div/div[2]/div[2]/div/div[2]/div/div[2]/div/div/div[2]/ul/li[4]/a/i')
        next_button.click()
    except:
        # Cuộn đến cuối trang web,
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Đợi cho trang web hiển thị dữ liệu trang tiếp theo
    wait_for_element_present(driver, By.CLASS_NAME, "z-paging-next")

    time.sleep(10)

    page_number += 1

# Đóng trình duyệt
    
time.sleep(5)
        
        
    
    
```
