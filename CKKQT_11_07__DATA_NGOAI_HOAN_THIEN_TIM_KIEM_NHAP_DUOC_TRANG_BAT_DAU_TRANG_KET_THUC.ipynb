```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC



def boc_tach_data(page, index_name=0):
    soup = BeautifulSoup(page, 'html.parser')
    table = soup.find_all(['table'])[3]
    soup2 = BeautifulSoup(str(table), 'html.parser')
    tr_array = soup2.find_all(['tr'])
    ketqua = []
    for i in tr_array:
        soup_tr = BeautifulSoup(str(i), 'html.parser')
        data_cot = []
        danh_sach_cot = soup_tr.find_all(['td'])
        for c in danh_sach_cot:
            data_cot.append(c.text)
        ketqua.append(data_cot)
    data = ketqua
    df = pd.DataFrame(data, columns=['STT', 'TÊN TRANG THIẾT BỊ Y TẾ', 'PHÂN NHÓM', 'CHỦNG LOẠI (MODEL)', 'NƯỚC SẢN XUẤT', 'HÃNG SẢN XUẤT', 'ĐƠN VỊ TÍNH', 'SỐ LƯỢNG', 'ĐƠN GIÁ TRÚNG THẦU (VNĐ)', 'SỐ QUYẾT ĐỊNH', 'ĐƠN VỊ ĐĂNG TẢI', 'NGÀY ĐĂNG TẢI'])
    # Xuất DataFrame thành file Excel
    df.to_excel(f"Data_ngoai_BVHNVD_{index_name}.xlsx", index=False)

   

    
    
    
# Khởi tạo trình điều khiển Firefox
driver = webdriver.Firefox()
# Mở trang web
url = "https://congkhaiketquathau.moh.gov.vn/Pages/admin/report/TT14/manager/mohVtIvdManager.zul?code=1"
driver.get(url)




# Chờ tải đầy đủ trang web (tuỳ thuộc vào trang web)
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ".row.row-space.z-div")))



# Tìm tất cả các phần tử có class "row row-space z-div"
elements = driver.find_elements(By.CSS_SELECTOR, ".row.row-space.z-div")

# Kiểm tra nếu có phần tử
if elements:
    # Truy cập vào từng phần tử và thực hiện các thao tác
    for index, element in enumerate(elements):
        # Tìm tất cả các phần tử input bên trong phần tử hiện tại
        input_elements = element.find_elements(By.CSS_SELECTOR, "input")

        # Chờ đến khi các phần tử input xuất hiện
        for input_element in input_elements:
            wait.until(EC.visibility_of(input_element))

        # Chờ thêm 1 giây để đảm bảo phần tử tương tác được
        time.sleep(1)

        # Thực hiện các thao tác tương ứng với từng hàng
        if index == 0:
            input_elements[0].clear()
            input_elements[0].send_keys("")  # NHẬP TÊN VẬT TƯ CẦN TÌM KIẾM,ĐANG BỎ TRỐNG
        elif index == 1:
            input_elements[0].clear()
            input_elements[0].send_keys("")  # NHẬP TÊN VẬT TƯ CẦN TÌM KIẾM,ĐANG BỎ TRỐNG
            input_elements[1].clear()
            input_elements[1].send_keys("Bệnh viện Hữu nghị Việt Đức")  # NHẬP TÊN BỆNH VIỆN CẦN LỌC, CHÍNH XÁC

            # CHỌN TÊN BỆNH VIỆN HIỆN RA. Click vào phần tử <i> để mở popup
            i_element = driver.find_element(By.CSS_SELECTOR, 'i.z-bandbox-icon.z-icon-search')
            i_element.click()
            # Chờ TÊN BỆNH VIỆN xuất hiện trong khoảng thời gian tối đa là 3 giây
            popup_div = WebDriverWait(driver, 3).until(
                EC.visibility_of_element_located((By.CSS_SELECTOR, 'div.z-bandbox-popup.z-bandbox-open.z-bandbox-shadow')))
            # Tìm và click vào TÊN BỆNH VIỆN
            div_element = popup_div.find_element(By.CSS_SELECTOR, 'div.z-listcell-content')
            div_element.click()
        elif index == 2:
            input_elements[0].clear()
            input_elements[0].send_keys("")  # NHẬP CHỦNG LOẠI, MÃ SẢN PHẨM,ĐANG BỎ TRỐNG
            input_elements[1].clear()
            input_elements[1].send_keys("")  # NHẬP HÃNG SẢN XUẤT, ĐANG BỎ TRỐTRỐNG
        elif index == 3:
            input_elements[0].clear()
            input_elements[0].send_keys("")  # NHẬP KHOẢNG GIÁ TRÚNG THẦU CẦN TÌM TỪ, ĐANG BỎ TRỐNG
            input_elements[1].clear()
            input_elements[1].send_keys("")  # NHẬP KHOẢNG GIÁ TRÚNG THẦU CẦN TÌM ĐẾN, ĐANG BỎ TRỐNG
            time.sleep(2)
        elif index == 4:
            # TÌM Ô NGÀY CÔNG KHAI
            # Tìm tất cả các phần tử input nested bằng class name
            nested_input_elements = driver.find_elements(By.CLASS_NAME, "z-datebox-input")

            # Lấy phần tử đầu tiên và thực hiện các thao tác
            nested_input_element1 = nested_input_elements[0]
            time.sleep(2)
            nested_input_element1.clear()
            time.sleep(2)
            nested_input_element1.send_keys("01/01/2022")

            # TÌM Ô ĐẾN NGÀY
            # Lấy phần tử thứ hai và thực hiện các thao tác
            nested_input_element2 = nested_input_elements[1]
            nested_input_element2.clear()
            nested_input_element2.send_keys("31/12/2022")

        time.sleep(5)

        
        
        
# Tìm nút "TÌM KIẾM" bằng CSS selector
button = driver.find_element(By.CSS_SELECTOR, 'button.btn-primary.z-button')
# Thực hiện click vào nút "TÌM KIẾM"
button.click()
# Chờ 30 giây để đợi kết quả tải và hiển thị
time.sleep(30)




def wait_for_element_present(driver, by, locator, timeout=10):
    wait = WebDriverWait(driver, timeout)
    wait.until(EC.presence_of_element_located((by, locator)))

    

    
# NHẬP SỐ TRANG VÀ TẢI LẠI DỮ LIỆU BẮT ĐẦU.

# Tìm và click vào phần tử ô input
input_element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div/div[2]/div[2]/div/div[2]/div/div[2]/div/div/div[2]/ul/li[3]/input')))
input_element.click()
time.sleep(5)
# Xoá nội dung ô nhập liệu
input_element.send_keys(Keys.BACKSPACE)
time.sleep(3)
# Nhập giá trị trang bắt đầu
input_element.send_keys("3") # TỚI ĐÂY DỮ LIỆU SẼ LOAD ĐẾN TRANG SỐ 3    
time.sleep(5)
# Giả lập nhấn phím Enter
input_element.send_keys(Keys.ENTER)
time.sleep(40)    
    
# TỚI ĐÂY DỮ LIỆU SẼ LOAD ĐẾN TRANG SỐ 3    
    
    
    
    
# NHẬP SỐ TRANG SẼ DỪNG LẠI  

end_page = 10 # NHẬP SỐ TRANG SẼ DỪNG LẠI
page_number = 1



while page_number <= end_page:
    print("Page", page_number)

    # Lấy dữ liệu từ trang hiện tại
    elem = driver.find_element(By.TAG_NAME, "body")
    data = elem.get_attribute('innerHTML')
    html_doc = str(data)
    boc_tach_data(html_doc, page_number)

    # Xử lý lỗi ElementNotInteractableException
    try:
        time.sleep(5)
        next_button = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div/div[2]/div[2]/div/div[2]/div/div[2]/div/div/div[2]/ul/li[4]/a/i')
        next_button.click()
    except:
        # Cuộn đến cuối trang web,
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    page_number += 1

    # Đợi cho trang web hiển thị dữ liệu trang tiếp theo
    wait_for_element_present(driver, By.CLASS_NAME, "z-paging-next")

    time.sleep(10)

# Đóng trình duyệt
driver.quit()

print("Done")
```
