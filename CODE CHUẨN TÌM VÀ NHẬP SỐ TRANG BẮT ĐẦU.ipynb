### CODE CHUẨN TÌM VÀ NHẬP SỐ TRANG BẮT ĐẦU, VÂN ĐỀ NẰM Ở THỜI GIAN CHỜ

```python
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.common.action_chains import ActionChains 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
import pandas as pd

# Khởi tạo trình điều khiển Firefox
driver = webdriver.Firefox()

# Thay đổi URL tại đây
url = "https://congkhaiketquathau.moh.gov.vn/Pages/admin/report/TT14/manager/mohVtIvdManager.zul?code=1"
driver.get(url)
time.sleep(10)

# CHỌN BỆNH VIỆN.
# Tìm danh sách các phần tử "đơn vị" bằng class name
don_vi_elements = driver.find_elements(By.CSS_SELECTOR, ".col-md-3.col-form-control.z-div input.z-bandbox-input")

# Lấy phần tử đầu tiên trong danh sách
don_vi_input = don_vi_elements[0]
# Xóa nội dung của ô input "đơn vị"
don_vi_input.clear()
# Gửi dữ liệu "Bệnh viện hữu nghị việt đức" vào trường input
don_vi_input.send_keys("Bệnh viện hữu nghị việt đức")

# Click vào phần tử <i> để mở popup
i_element = driver.find_element(By.CSS_SELECTOR, 'i.z-bandbox-icon.z-icon-search')
i_element.click()

# Chờ phần tử popup xuất hiện trong khoảng thời gian tối đa là 3 giây
popup_div = WebDriverWait(driver, 3).until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'div.z-bandbox-popup.z-bandbox-open.z-bandbox-shadow')))

# Tìm và click vào phần tử trong popup
div_element = popup_div.find_element(By.CSS_SELECTOR, 'div.z-listcell-content')
div_element.click()
time.sleep(5)

# Tìm danh sách các phần tử "ngày công khai" bằng class name
ngay_cong_khai_elements = driver.find_elements(By.CLASS_NAME, "z-datebox-input")
# Lấy phần tử thứ 1 trong danh sách
ngay_cong_khai = ngay_cong_khai_elements[0]
ngay_cong_khai.clear()
time.sleep(5)

# Thay đổi ngày công khai tại đây
ngay_cong_khai.send_keys("01/01/2022")
time.sleep(5)

# Tìm danh sách các phần tử "Đến ngày" bằng class name
den_ngay_elements = driver.find_elements(By.CLASS_NAME, "z-datebox-input")
# Lấy phần tử thứ 2 trong danh sách
den_ngay = den_ngay_elements[1]
den_ngay.clear()

# Thay đổi ngày đến tại đây
den_ngay.send_keys("31/12/2022")
time.sleep(5)


# Tìm nút (button TÌM KIẾM) bằng CSS selector
button = driver.find_element(By.CSS_SELECTOR, 'button.btn-primary.z-button')
# Thực hiện click vào nút TÌM KIẾM
button.click()
time.sleep(15)


# Tìm và click vào phần tử ô input TRANG BẮT ĐẦU
input_element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div/div[2]/div[2]/div/div[2]/div/div[2]/div/div/div[2]/ul/li[3]/input')))
input_element.click()
time.sleep(5)

# Xoá nội dung ô nhập liệu
input_element.send_keys(Keys.BACKSPACE)
time.sleep(3)

# Nhập giá trị trang bắt đầu
input_element.send_keys("3")
time.sleep(5)
# Giả lập nhấn phím Enter
input_element.send_keys(Keys.ENTER)
time.sleep(5)

# Các bước còn lại của mã của bạn
# ...
# Cuộn đến cuối trang web, CLICK VÀO NÚT NEXT THEO XPATH
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)
input_element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div/div/div/div[2]/div[2]/div/div[2]/div/div[2]/div/div/div[2]/ul/li[4]/a/i')))
input_element.click()
```
