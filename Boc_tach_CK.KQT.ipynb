```python
tr_array = soup2.find_all(['tr'])
tr_array   # in ra xem thử
```

```python
for tr in tr_array:
    print(tr)
```

```python
ketqua = []
for i in tr_array:
    soup_tr = BeautifulSoup(str(i), 'html.parser')
    data_cot = []
    danh_sach_cot = soup_tr.find_all(['td'])
    for c in danh_sach_cot:
        data_cot.append(c.text)
    ketqua.append(data_cot)
print(ketqua)
```

```python
# in thử 1 dòng data đầy đủ
for r in ketqua:
    data_r = ",".join(r) # mỗi 1 cột cách nhau bằng dấu ,
    print(data_r)
    break    # in 1 dòng , tới đây là dòng có đầy đủ data
```

```python
# mở file data.csv, ghi dữ liệu vào
f = open("data.csv", "a", encoding="utf-8") # # mở file csv có tên data, theo kiểu a, cần có encoding="utf-8"
#f = open("data.csv", "a")  # nếu chạy lệnh này sẽ gặp lỗi UnicodeEncodeError: 'charmap' codec can't encode chara
# là không đúng fomat, thêm encoding="utf-8" vào cuối là được
for r in ketqua:
    data_r = ",".join(r)
    f.write(data_r)   
f.close()    # tới đây đã ghi ra file data.csv
```

```python
def boc_tach_data(page, index_name=0):
    soup = BeautifulSoup(page, 'html.parser')
    table = soup.find_all(['table'])[1]
    soup2 = BeautifulSoup(str(table), 'html.parser')
    tr_array = soup2.find_all(['tr'])
    ketqua = []
    for i in tr_array:
        soup_tr = BeautifulSoup(str(i), 'html.parser')
        data_cot = []
        danh_sach_cot = soup_tr.find_all(['td'])
        for c in danh_sach_cot:
            data_cot.append(c.text)
        ketqua.append(data_cot)
    import pandas as pd

    data = ketqua

    df = pd.DataFrame(data)

    # Xuất DataFrame thành file CSV
    df.to_csv(f"data_{index_name}.csv", index=False)

    
```

```python
boc_tach_data(str(soup))

```
